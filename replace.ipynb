```python
#Replace encoding using loop
df1=sns.load_dataset("titanic")
```

```python
from pandas.api.types import is_numeric_dtype
```

```python
for col in df1.columns:
    if is_numeric_dtype(df1[col]):
        continue
    else:
        uni = df1[col].nunique()
        val = df1[col].unique().tolist()
        if uni == 2:
            df1[col] = df1[col].replace(val, [0, 1])
        elif uni==3:
            df1[col]= df1[col].replace(val, [0,1,2])
        elif uni==4:
            df1[col]= df1[col].replace(val,[0,1,2,3])
```
